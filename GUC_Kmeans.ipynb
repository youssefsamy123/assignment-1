> ### 1.1. Code for the distance function 

write a function to calculate the distance between a set of cluster centroids and all the data points 

Cluster_Distances = GUC_Distance ( Cluster_Centroids, Data_points, Distance_Type) 
* The function shoud be able to handel data points with any number of features and any number of points 
* The function should be able to handle any K number of cluster centroids 
* To inrease speed do that in matrix format using numpy
* This should produce a matrix with rows equal to data points columns equal K 
* You should be able to calculate different types of Distance. In this assginmnet you are required to implement the following two 
    - Ecluidian distance
    - Pearson correlation distance
  

```python
def GUC_Distance(Cluster_Centroids, Data_points, Distance_Type):
    if Distance_Type == 'euclidean':
        
        Cluster_Distance = np.linalg.norm(Data_points[:, np.newaxis, :] - Cluster_Centroids, axis=2)
    elif Distance_Type == 'pearson':
        Cluster_Distance = np.zeros((Data_points.shape[0], Cluster_Centroids.shape[0]))

        for i in range(Data_points.shape[0]):
            for j in range(Cluster_Centroids.shape[0]):
                centered_data = Data_points[i] - np.mean(Data_points[i])
                centered_centroid = Cluster_Centroids[j] - np.mean(Cluster_Centroids[j])
                numerator = np.dot(centered_data, centered_centroid)
                denominator = np.linalg.norm(centered_data) * np.linalg.norm(centered_centroid)
                pearson_correlation = 1 - (numerator / denominator)
                Cluster_Distance[i, j] = pearson_correlation

    else:
        raise ValueError("Invalid Distance_Type. Use 'euclidean' or 'pearson'.")

    return Cluster_Distance
```

> ### 1.2. Code for K means function 

> Loop until a stopping condition 

> > in the loop do the following 
> > 1. Cluster Assginmnet 
> > > * Use the GUC_Distance to calculate between each cluster head and all the points 
> > > * Find the value and index of the minimum distance  for each row. 
> > > * for each point in the row use the index of the minmum distance to represent the cluster Centroid that this point will belong to.
> > > * Calculate and record the mean square distance for each cluster  
> > > * Use the mean square distances to calculate a stopping condition (e.g the sum of mean distances)
> > 2. Updtate Centroids  
> > > * Update the value for the cluster heads by calculating the mean value for the points in each cluster 
> > >  * Calculate the Cluster Metric (Distortion Function)
> > > > a number indicating the how good the clusters are . An example cluster metric is the  (sum of squared error between each point and its cluster center) 

```python
def GUC_Kmean(Data_points, Number_of_Clusters, Distance_Type, max_iterations=100, tolerance=1e-4):
    
    num_points, num_features = Data_points.shape
    min_values = np.min(Data_points, axis=0)
    max_values = np.max(Data_points, axis=0)
    Cluster_Centroids = np.random.uniform(min_values, max_values, size=(Number_of_Clusters, num_features))

    
    iteration = 0
    prev_cluster_metric = np.inf

    while iteration < max_iterations:
        
        Cluster_Distance = GUC_Distance(Cluster_Centroids, Data_points, Distance_Type)
        Assigned_Clusters = np.argmin(Cluster_Distance, axis=1)
        Cluster_Metric = np.sum(np.min(Cluster_Distance, axis=1))

        
        New_Cluster_Centroids = np.zeros_like(Cluster_Centroids)
        for i in range(Number_of_Clusters):
            New_Cluster_Centroids[i] = np.mean(Data_points[Assigned_Clusters == i], axis=0)

        
        if np.abs(prev_cluster_metric - Cluster_Metric) < tolerance:
            break

        
        Cluster_Centroids = New_Cluster_Centroids
        prev_cluster_metric = Cluster_Metric
        iteration += 1

    return Assigned_Clusters, Cluster_Centroids   
```

```python
# helper function that allows us to display data in 2 dimensions an highlights the clusters
def display_cluster(X,km=[],num_clusters=0):
    color = 'brgcmyk'  #List colors
    alpha = 0.5  #color obaque
    s = 20
    if num_clusters == 0:
        plt.scatter(X[:,0],X[:,1],c = color[0],alpha = alpha,s = s)
    else:
        for i in range(num_clusters):
            plt.scatter(X[km.labels_==i,0],X[km.labels_==i,1],c = color[i],alpha = alpha,s=s)
            plt.scatter(km.cluster_centers_[i][0],km.cluster_centers_[i][1],c = color[i], marker = 'x', s = 100)
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
def plot_cluster_performance(X, title, Distance_Type):
    silhouette_scores = []
    for n_clusters in range(2, 11):
        Assigned_Clusters, _ = GUC_Kmean(X, n_clusters, Distance_Type)
        
        
        if len(np.unique(Assigned_Clusters)) < 2:
            silhouette_scores.append(0)
        else:
            silhouette_scores.append(silhouette_score(X, Assigned_Clusters))
    
    plt.plot(range(2, 11), silhouette_scores, marker='o')
    plt.title(title)
    plt.xlabel('Number of Clusters')
    plt.ylabel('Silhouette Score')
    plt.show()


plt.subplot(1, 3, 3)
plt.title('Moons Data')
n_samples = 1000
X_moons, _ = make_moons(n_samples=n_samples, noise=0.1)
plot_cluster_performance(X_moons, title='Moons Data', Distance_Type='pearson')
```
